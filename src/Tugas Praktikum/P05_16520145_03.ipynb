```python
# 1. Line chart untuk harga BTC setiap harinya. Gunakan nilai close sebagai nilai harian
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("crypto.csv")
btc = df.loc[df["name"] == "BTC"] # mengambil data BTC

btc.plot(kind = "line", x = "date", y = "close")
```

```python
# 2. Scatter plot dengan nilai high sebagai sumbu x dan nilai low sebagai sumbu y untuk koin ETH
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("crypto.csv")
eth = df.loc[df["name"] == "ETH"] # mengambil data ETH

eth.plot(kind = "scatter", x = "high", y = "low")
```

```python
# 3. Pie chart banyaknya data dari masing-masing koin
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("crypto.csv")

df["name"].value_counts().plot(kind = "pie")
```

```python
# 4. Pie chart nilai kapitalisasi pasar semua mata uang pada tanggal 4 Juli 2016
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("crypto.csv")
pie = df.loc[df["date"] == "2016-07-04"] # mengambil data 4 Juli 2016

pie[["name", "market"]].value_counts().plot(kind = "pie")
```

```python
# 5. Line chart untuk membandingkan nilai high BTC, ETH, XRP, USDT dan XMR
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("crypto.csv")

df.plot(kind = "line", y = "high", x = "name")
```
