```python
# 1. Histogram distribusi nilai kalkulus
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("nilai_uas.csv")

df["nilai_kal"].plot(kind = "hist", bins = [0,20,40,60,80,100])
```

```python
# 2. Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("nilai_uas.csv")

df["fakultas"].value_counts().plot(kind = "barh")
```

```python
# 3. Diagram pie banyaknya mahasiswa masing-masing fakultas
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("nilai_uas.csv")

df["fakultas"].value_counts().plot(kind = "pie")
```

```python
# 4. Fakultas dengan mahasiswa terbanyak adalah STEI
# Diagram yang lebih baik dalam menampilkan mahasiswa terbanyak adalah diagram batang horizontal
# karena pada diagram batang horizontal, sumbu x menunjukkan jumlah mahasiswa
# sehingga dari panjang batang tersebut dapat diperkirakan jumlah mahasiswanya
# sedangkan pada pie tidak terdapat angka, akibatnya sulit untuk mengetahui jumlah mahasiswanya
```

```python
# 5. Scatter plot dengan nilai kimia sebagai x dan fisika sebagai y
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("nilai_uas.csv")

df.plot(kind = "scatter", x = "nilai_kim", y = "nilai_fis")
```
